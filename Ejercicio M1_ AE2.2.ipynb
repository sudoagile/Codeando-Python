```python
año1 = int(input("Año 1:"))
año2 = int(input("Año 2:"))


años = list (range(año1, año2+1))
contador = 0
for i in años:
  if i % 4 == 0 and (i % 100 != 0 or i % 400 == 0):
    contador += 1
    print (i,"Es bisiesto")

print("Existen: ", contador , " años biciestos")

#if año % 4 == 0 and (año % 100 != 0 or año % 400 == 0):
#	print("Es bisiesto")
#else:
#	print("No es bisiesto")
```
